```python
# Filtering the original dataset in order to access the desired columns for the various financial services accessed
non = data[(data.mobile_money_classification == 0) | (data.mobile_money_classification == 1)]

mobile = data[(data.mobile_money_classification == 2)]

both = data[(data.mobile_money_classification == 3)]

# Filtering the categories to be presented against the financial services accesed
age = data.Q1
gender = data.Q2
marital_status = data.Q3
land_ownership = data.Q5
```

```python
# Age vs Financial services plots
age_non = non.Q1
age_mobile = mobile.Q1
age_both = both.Q1

plt.hist(age_non, color = ['purple'])
plt.xlabel('Age distribution')
plt.ylabel('Number of people')
plt.title('The age distribution of people who do not mobile money')
plt.show()

plt.hist(age_mobile, color = ['plum'])
plt.xlabel('Age distribution')
plt.ylabel('Number of people')
plt.title('The age distribution of people who only use mobile money')
plt.show()

plt.hist(age_both, color = ['indigo'])
plt.xlabel('Ages')
plt.ylabel('Number of people')
plt.title('The age distribution of people who use mobile money and one other financial service')
plt.show()
```

```python
# Gender vs Financial services plot 
gender_non = non.Q2
gender_mobile = mobile.Q2
gender_both = both.Q2

g_non = gender_non.groupby(gender_non.iloc[:-1]).count()
gender_non_list = list('')
for i in g_non :
    gender_non_list.append(i)
labels = ['Male','Female']

plt.bar(labels,gender_non_list, color = ['green', 'yellow'])
plt.xlabel('Gender')
plt.ylabel('Number of people')
plt.title('Gender categories who do not use mobile money')
plt.show()

g_mobile = gender_mobile.groupby(gender_mobile.iloc[:-1]).count()
gender_mobile_list = list('')
for i in g_mobile :
    gender_mobile_list.append(i)
labels = ['Male','Female']

plt.bar(labels,gender_mobile_list, color = ['orange', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Number of people')
plt.title('Gender categories who use mobile money')
plt.show()

g_both = gender_both.groupby(gender_both.iloc[:-1]).count()
gender_both_list = list('')
for i in g_both :
    gender_both_list.append(i)
labels = ['Male','Female']

plt.bar(labels,gender_both_list, color = ['red', 'blue'])
plt.xlabel('Gender')
plt.ylabel('Number of people')
plt.title('Gender categories who use mobile money and at least one other financial service')
plt.show()
```

```python
# Marital status vs Financial sevices plot
marital_status_non = non.Q3
marital_status_mobile = mobile.Q3
marital_status_both = both.Q3

ms_non = marital_status_non.groupby(marital_status_non.iloc[ :-1]).count()
ms_non_empty_list = list('')
labels = ['married', 'divorced', 'widowed', 'single/never_married']
for i in ms_non :
    ms_non_empty_list.append(i)  
    
plt.bar(labels,ms_non_empty_list, color = ['pink','orange', 'yellow', 'green'])
plt.xlabel('Marital status')
plt.ylabel('Number of people')
plt.title('Marital statuses of people who do not use mobile money')
plt.show()  

ms_mobile = marital_status_mobile.groupby(marital_status_mobile.iloc[ :-1]).count()
ms_mobile_empty_list = list('')
labels = ['married', 'divorced', 'widowed', 'single/never_married']
for i in ms_mobile :
      ms_mobile_empty_list.append(i)
    
plt.bar(labels,ms_mobile_empty_list, color = ['pink','orange', 'yellow', 'green'])
plt.xlabel('Marital status')
plt.ylabel('Number of people')
plt.title('Marital statuses of people who use mobile money')
plt.show()  

ms_both = marital_status_both.groupby(marital_status_both.iloc[ :-1]).count()
ms_both_empty_list = list('')
labels = ['married', 'divorced', 'widowed', 'single/never_married']
for i in ms_both :
    ms_both_empty_list.append(i)
    
plt.bar(labels,ms_both_empty_list, color = ['pink','orange', 'yellow', 'green'])
plt.xlabel('Marital status')
plt.ylabel('Number of people')
plt.title('Marital statuses of people who use mobile money and at least one financial service')
plt.show()  
```

5. Conclusion and insoghts drawn from the data above

From the graphs visualizing the age vs mobile money usage, we can see that young people between the ages of 10 to 30 make up the largest percentage of people who use mobile money. The highest group of users for mobile money and one other finacial service range between ages 25 to 30. However young people aged between 10 and 40 years old still made up the largest segment of people not using any financial services in 2012.

The graphs depicting gender vs the use of mobile money show us that women are the highest gender category for the no financial services as well as for mobile money. Men rank higher with the use of both mobile money and at least one other financial service use. What this tells us is that women may still have been the most disadvantaged when it comes to access to financial services in Tanzania in 2012. This could directly affect the growth rate of the country as many women head up households and yet cannot access funds to improve their lives and the lives of their dependents. 

The recurring theme observed with the graphs showing the the marital status of the population vs the use of mobile money are that married people seem to be the majority useres across all financial services. This alludes to the possibility that married people also make up the the largest population within the category of marital status. As observed above, young people between the ages of 10 - 40 years of age make up the largest part of the population. Combining this detail together with the marital status information tells us that there is also a large number of young people who were married in Tanzania in 2012.The single/never married category comes in second highest across all financial services use. 

The land ownership vs mobile money plots depict that across the board, people who personally own the land they live on, are active within with financial services sector. However, the number of people who own land and do not use any financial services is still the highest. This means people own assets with which to participate ( whether it be through leasing their property or even using it as surety for a potential business) in the economy but might not due to access to financial services.
